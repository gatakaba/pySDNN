```python
import numpy as np
import matplotlib.pyplot as plt
from pysdnn import coding
plt.rcParams["figure.figsize"] = [12,8]
```

# codingクラスの使い方

```python
# コードパターンの要素数
code_pattern_dim=8
# 実数の分割数
input_division_num=10
# 反点数
reversal_num=1
# 入力データの次元数
input_dim=2
# インスタンスの生成
pc = coding.PatternCoding(code_pattern_dim,input_division_num,reversal_num,input_dim)
```

```python
x = np.array([0.5,0.5])
y = np.array([0.6,0.6])

pattern_x=pc.coding(x)
pattern_y=pc.coding(y)

print("pattern_x:"+str(pattern_x))
print("pattern_y:"+str(pattern_y))

print("pattern_y - pattern_x:" + str(pattern_x-pattern_y))
```

# 入力とコードパターン

```python
code_pattern_dim = 200
input_division_num = 100
reversal_num = 1
input_dim = 1

pc = coding.PatternCoding(code_pattern_dim,input_division_num,reversal_num,input_dim)
```

```python
X=np.linspace(0,1,100)
pattern_list=[]
for x in X:
    pattern_list.append(pc.coding(np.array(x)))

plot_x=X
plot_y=range(200)

plt.xlabel("x")
plt.ylabel("code pattern vector")
plt.pcolor(plot_x,plot_y,np.array(pattern_list).T)
plt.colorbar()
```

## 入力が[0,1]外に及ぶ場合の挙動

```python
X=np.linspace(-10,10,100)
pattern_list=[]
for x in X:
    pattern_list.append(pc.coding(np.array(x)))

plot_x=X
plot_y=range(200)

plt.xlabel("x")
plt.ylabel("code pattern vector")
plt.pcolor(plot_x,plot_y,np.array(pattern_list).T)
plt.colorbar()
```

## スケーリングを行った場合(pc.codingの引数を設定)

```python
X=np.linspace(-10,10,100)
pattern_list=[]
for x in X:
    pattern_list.append(pc.coding(np.array(x),-10,10))

plot_x=X
plot_y=range(200)

plt.xlabel("x")
plt.ylabel("code pattern vector")
plt.pcolor(plot_x,plot_y,np.array(pattern_list).T)
plt.colorbar()
```
